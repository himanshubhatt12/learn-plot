# Line Graph

<i>Wikipedia says that a line graph is:</i>
<p>A line chart or line plot or line graph or curve chart is a type of chart which displays information as a series of data points called 'markers' connected by straight line segments.</P>

<p>We use these graphs to plot functions or relationship between two variables.</p>
<p>For ex. price v/s quality of a product</p>

```
x = [1, 2, 3, 4, 5, 6, 7]
y = [8, 4, 6, 4, 9, 4, 2]
```

```
plt.figure(figsize = (9, 7)) 
plt.plot(x, y, marker = 'o', markersize = 12)
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("Plot with markers")
```

```
plt.plot(x, y, 'yo')
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("Plot with yellow markers")
```

# Annotations
Text to give some more information about a point

```
plt.plot(x, y, 'b')     # 'b' specifies blue
plt.xlabel("x value")   # sets the label for x-axis of the plot
plt.ylabel("y value")  # set the label for y-axis of the plot
plt.text(5, 9, "Max y") # (5, 9) is the data point to annotate
plt.title("Plot x y")  # sets the title of plot
```

<h3>Two plots in a graph(used for comparison)</h3>

<p>Suppose we are having 2 companies and their sales data for a week and we want to compare the performance of these companies.</p>
<p>In this case it seems natural to use two line charts in a single figure</p>

```
days = ["Sun", "Mon", "Tue", "Wed", "Thu", "Fri", "Sat"]
company1 = [50, 40, 23, 34, 20, 10, 50]
company2 = [30, 40, 30, 45, 20, 45, 30]
plt.figure(figsize = (10, 8))
plt.plot(days, company1) # plots line chart for company 1
plt.plot(days, company2) # plots line chart for company 2
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales data of company 1 and company 2 for this week")
```

<h3>Using legends</h3>
<p>There is a problem with the above plot.What's that?</p>
<p>With above graph we cannot say which colour is for which company</p>
<p>So we are using legends to differentiate it</p>

```
days = ["Sun", "Mon", "Tue", "Wed", "Thu", "Fri", "Sat"]
company1 = [50, 40, 23, 34, 20, 10, 50] # sales of company1
company2 = [30, 40, 30, 45, 20, 45, 30] # sales of company 2
plt.figure(figsize = (10, 8))
plt.plot(days, company1, label = "Company 1") # plots line chart for company 1
plt.plot(days, company2, label = "Company 2") # plots line chart for company 2
plt.xlabel("Days")
plt.ylabel("Sales")
plt.legend() # using legend to differentiate the graphs of two companies
plt.title("Sales data of company 1 and company 2 for this week")
```

# Using subplots 

```
days = ["Sun", "Mon", "Tue", "Wed", "Thu", "Fri", "Sat"]
company1 = [50, 40, 23, 34, 20, 10, 50] # sales of company1
company2 = [30, 40, 30, 45, 20, 45, 30] # sales of company 2
fig, ax = plt.subplots(nrows = 2, ncols = 1, figsize = (13, 15))
ax[0].plot(days,company1, color = "g")
ax[1].plot(days, company2, color = "b")
ax[0].set_xlabel("Days") # method set_xlabel, set_ylabel are used when working with subplots
ax[0].set_ylabel("Sales")
ax[1].set_xlabel("Days")
ax[1].set_ylabel("Sales")
ax[0].set_title("Company 1") # method set_title is used when working with subplots
ax[1].set_title("Company 2")
```

```
days = ["Sun", "Mon", "Tue", "Wed", "Thu", "Fri", "Sat"]
company1 = [50, 40, 23, 34, 20, 10, 50] # sales of company1
company2 = [30, 40, 30, 45, 20, 45, 30] # sales of company 2
fig, ax = plt.subplots(nrows = 2, ncols = 1, figsize = (14, 12), sharex = True)
ax[0].plot(days,company1, color = "g")
ax[1].plot(days, company2, color = "b")
ax[0].set_ylabel("Sales") # method set_xlabel, set_ylabel are used when working with subplots
ax[1].set_xlabel("Days")
ax[1].set_ylabel("Sales")
ax[0].set_title("Company 1") # method set_title is used when working with subplots
ax[1].set_title("Company 2")
```

# Bar graph

```
categories = ["WebDev", "DataScientist", "AndroidDev"]
```

```
values = [40, 30, 30]
```

```
plt.bar(categories, values, color = "rebeccapurple", width = 0.8)
# categories is the x parameter
# values are the height parameters
plt.title("Software Job roles in a company\nFor the year 2020")
plt.xlabel("Job role")
plt.ylabel("Number of software Engineers")
```

# Double Bar Graph

<p>Suppose there are 2 malls and they we want to see their sales in three different items(in millions):
<ul>
  <li>Electronics</li>
  <li>Clothing</li>
  <li>Kitchen Accessories</li>
</ul>


```
categories = ["Electronics", "Clothing", "Kitchen Accessories"] 
mall1 = [50, 30, 20]
mall2 = [40, 60, 30]
fig, ax = plt.subplots(2,1)
ax[0].bar(categories, mall1, color = "b", label = "Mall 1")  # here x = categories and height = mall1
ax[1].bar(categories, mall2, color = "g", label = "Mall 2")
ax[0].legend()
ax[1].legend()
plt.show()
```
